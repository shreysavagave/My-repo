```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

```python
df=df.drop(columns =['PassengerId','SibSp','Name','Parch','Ticket','Cabin','Embarked'])
```

```python
df['survived'] = df['Survived']
```

```python
df = df.drop('Survived',axis=1)
```

```python
target = df.survived
inputs = df.drop('survived',axis=1)
```

```python
inputs.isnull().sum()
```

```python
inputs['Sex'].unique()
```

```python
inputs['Sex'] = inputs['Sex'].map({'male':1 ,'female':0})
```

```python
inputs.head(3)
```

```python
inputs['Age'] = inputs['Sex'].fillna(inputs['Sex'].median)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trian , X_test , y_trian, y_test = train_test_split(inputs,target,test_size=0.33)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_trian,y_trian)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
